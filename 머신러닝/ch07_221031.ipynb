## 텍스트 정규화
- 문장에 URL과 구두점처럼 분석에 불필요한 문자가 많음
- Cleansing 할 필요 있음.
  + 정규표현식을 통해 처리.
- 정규표현식
  + [0-9] : 숫자 0-9 중에 한 문자와 일치한다. 
    + [0-9 a-z] : 숫자 0-9, 소문자 a-z중에 한 문자 이상이 일치


```python
text
```

```python
import re

# 숫자 0-9, 영문 대소문자, URL포함 시, 해당 문자열 제거
reg_text = re.sub(r'[0-9a-zA-Z]+', '', text)
reg_text = re.sub(r'[:;/+\.-]', '', reg_text)

# 정규표현식을 사용해 줄바꿈, 공백 제거하라
reg_text = re.sub(r'[\s\n■]', '', reg_text)
reg_text
```

```python
for token in okt.pos(reg_text):
  print(token)
```

## 머신러닝을 위한 데이터 준비
- 모든 기사에 대한 형태소 분석 작업 진행

```python
import os
import re
from konlpy.tag import Okt
from tqdm import tqdm
okt = Okt()

#기사 폴더 지정
DATA_PATH = '/content/drive/MyDrive/Colab Notebooks/산대특/nlp/ch07/data/'
dirs = ['it-life-hack', 'movie-enter']

# 단어와 레이블을 보존하는 리스트 생성
docterm = [] # 설명변수
label = [] # 반응(=종속/목적)변수
tmp1 = []
tmp2 = ''

#각 폴더의 파일을 하나씩 읽어오기.
for i, d in enumerate(tqdm(dirs)):

  #파일 취득
  files = os.listdir(DATA_PATH + d)
  #print(files)

  for file in files:
    f = open(DATA_PATH + d +'/'+ file, 'r' , encoding='utf-8')
    text = f.read()
    #print(text)

    #정규표현식에서 불필요한 문자열을 제거해서 표시
    # 숫자 0-9, 영문 대소문자, URL포함 시, 해당 문자열 제거
    reg_text = re.sub(r'[0-9a-zA-Z]+', '', text)
    reg_text = re.sub(r'[:;/+\.-]', '', reg_text)

    # 정규표현식을 사용해 줄바꿈, 공백 제거하라
    reg_text = re.sub(r'[\s\n■]', '', reg_text)
    # reg_text

    # 명사로 필터링된 형태소 분석
    for token in okt.nouns(reg_text):
      # print(token)
      tmp1.append(token)
      tmp2 = ' '.join(tmp1)
    #print(tmp1)
    
    #단어 리스트를 docterm에 추가한다.
    docterm.append(tmp2)
    
    #초기화 
    tmp1=[]

    #기사 표시
    label.append(i)

    f.close()
```

## 카운트 기반의 벡터화
- 교재 205p

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
docterm_cv = cv.fit_transform(np.array(docterm))
docterm_cnt = docterm_cv.toarray()

pd.DataFrame(docterm_cnt).head()
```

```python
print(pd.DataFrame(docterm_cnt).shape)
```

```python
print(len(cv.get_feature_names()))
```

- 전체 기사에 대해 단어의 발생횟수를 카운트하고 발생횟수가 가장 많은 순서로 5개의 단어 표시하기

```python
wcnt = []

#단어 출현수 카운트
docterm_wcnt = np.sum(a = docterm_cnt, axis = 1)

#단어와 그 출현수를 하나씩 취득
for w, cnt in zip(cv.get_feature_names(), docterm_wcnt):
  wcnt.append([w, cnt])

#데이터 프레임으로 변환, 내림차순 정렬
wcnt_df = pd.DataFrame(wcnt).sort_values(1, ascending=False)
wcnt_df.head()
```

## TF-IDF 계산
